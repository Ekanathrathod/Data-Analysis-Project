```python
# Observed data: Satisfaction vs. Class
# Format: [[Business Dissatisfied, Economy Dissatisfied, Economy Plus Dissatisfied],
#          [Business Satisfied, Economy Satisfied, Economy Plus Satisfied]]
obs = np.array([[18013, 35219, 5373], [43977, 22898, 4007]])

# Perform Chi-square test
chi2, p_val, degree_of_freedom, expected = chi2_contingency(obs)

# Output results
print('Chi-Square Test Results for Class and Satisfaction:')
print(f'Chi-square Statistic: {chi2:.4f}')
print(f'p-value: {p_val:.4f}')
print(f'Degrees of Freedom: {degree_of_freedom}')
print('Expected Frequencies:\n', expected)

# Interpretation
alpha = 0.05
if p_val < alpha:
    print('Result: Significant association between Class of travel and Passenger Satisfaction (Reject H0)')
else:
    print('Result: No significant association between Class of travel and Passenger Satisfaction (Fail to Reject H0)')
```

### Step-by-Step explanation of the Chi-square test for the Passenger Satisfaction and Class of Travel :
#### Problem Statement:
* We are analyzing whether there is a significant association between the class of travel (Business, Economy, Economy Plus) and passenger satisfaction (Dissatisfied, Satisfied) using the Chi-square test for independence.

#### Hypotheses:
* Null Hypothesis (H0): There is no association between the class of travel and passenger satisfaction. The variables are independent.
* Alternative Hypothesis (H1): There is an association between the class of travel and passenger satisfaction. The variables are not independent.

#### Chi-Square Test for Independence:
* The Chi-square test evaluates whether the observed frequencies in the contingency table differ significantly from the expected frequencies under the  null hypothesis.
* Chi-square statistic (): Measures how much the observed frequencies deviate from the expected frequencies.  
* p-value: Determines the probability of observing the data if the null hypothesis is true.
* Degrees of Freedom (df): Calculated as (1)  (1) (rows1)  (columns1) = (21)(31) = 2(21)(31)=2.

#### Calculations:
* The test uses the observed values and calculates the expected frequencies assuming the null hypothesis is true.

#### Results from the Code:
* Chi-square Statistic: A high value indicates a greater discrepancy between observed and expected frequencies.
* p-value: A low p-value (typically < 0.05) suggests rejecting the null hypothesis.
* Expected Frequencies: The matrix of expected values under the assumption of independence.

#### Interpretation:
* If p < 0.05, we reject the null hypothesis, indicating a significant association between class and satisfaction.
* If p  0.05, we fail to reject the null hypothesis, suggesting no significant association.

#### Conclusion:
Based on the test results:
* Significant Association: If the p-value is low, the class of travel impacts passenger satisfaction.
* No Association: If the p-value is high, satisfaction levels are independent of travel class.

```python
# Observed data: Satisfaction vs. Gender
# Format: [[Female Dissatisfied, Female Satisfied], [Male Dissatisfied, Male Satisfied]]
obs = np.array([[22904, 42799], [35701, 28083]])

# Perform Chi-square test
chi2, p_val, degree_of_freedom, expected = chi2_contingency(obs)

# Output results
print('Chi-Square Test Results:')
print(f'Chi-square Statistic: {chi2:.4f}')
print(f'p-value: {p_val:.4f}')
print(f'Degrees of Freedom: {degree_of_freedom}')
print('Expected Frequencies:\n', expected)

# Interpretation
alpha = 0.05
if p_val < alpha:
    print('Result: Significant association between Gender and Satisfaction of passenger (Reject H0)')
else:
    print('Result: No significant association between Gender and Satisfaction of passenger (Fail to Reject H0)')
```

##  Point-by-Point explanation of the Chi-square test for the association between Gender and Passenger Satisfaction:
#### Objective:
* To determine if there is a significant association between passenger gender (Female, Male) and satisfaction (Dissatisfied, Satisfied) using a Chi-square test for independence.

#### Dataset Overview:

* Total Male Passengers: 63,784
* Total Passengers: 129,487

#### Hypotheses:
Null Hypothesis (H0):
* There is no association between gender and satisfaction. The variables are independent.
Alternative Hypothesis (H1):
* There is an association between gender and satisfaction. The variables are not independent

#### Statistical Test: Chi-Square Test for Independence
Purpose:
* The Chi-square test evaluates whether the observed frequencies in the contingency table differ significantly from expected frequencies under the null hypothesis.

#### Results Interpretation:
* Chi-square Statistic: Measures the discrepancy between observed and expected frequencies.
* p-value: If p < 0.05, reject the null hypothesis.
* Significance Level (): 0.05
  
Decision Rule:
* p < 0.05: Significant association between gender and satisfaction.
* p  0.05: No significant association.
